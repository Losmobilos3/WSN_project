```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class EnergestStat:
    def __init__(self, cpu, lpm, deep_lpm, total, RX, TX, OFF):
        self.cpu = cpu
        self.lpm = lpm
        self.deep_lpm = deep_lpm
        self.total = total
        self.RX = RX
        self.TX = TX
        self.OFF = OFF


class Experiment:
    class Instance:
        def __init__(self, n_acks, n_attempts, n_relay_acks, n_relay_attempts, x_labels, energest_stats):
            self.n_acks = n_acks
            self.n_attempts = n_attempts
            self.n_relay_acks = n_relay_acks
            self.n_relay_attempts = n_relay_attempts
            self.n_total_acks = n_acks + n_relay_acks
            self.n_total_attempts = n_attempts + n_relay_attempts
            self.x_labels = x_labels
            self.energest_stats = energest_stats
    
    def __init__(self, instances):
        self.instances = instances
        self.n_acks = np.array([inst.n_acks for inst in instances])
        self.n_attempts = np.array([inst.n_attempts for inst in instances])
        self.n_relay_acks = np.array([inst.n_relay_acks for inst in instances])
        self.n_relay_attempts = np.array([inst.n_relay_attempts for inst in instances])
        self.n_total_acks = self.n_acks + self.n_relay_acks
        self.n_total_attempts = self.n_attempts + self.n_relay_attempts
        self.x_labels = np.array([inst.x_labels for inst in instances])


    def plot(self):
        print(self.n_total_acks.shape, self.n_total_attempts.shape, self.x_labels.shape)
        plt.plot(self.x_labels, self.n_total_acks / self.n_total_attempts, label='Total ACKs/Attempts')
```
